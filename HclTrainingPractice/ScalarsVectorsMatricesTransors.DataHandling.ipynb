```python
#import
import numpy as np
```

```python
#scalar
scalar = np.array(5)
print('Scalar : ', scalar)
print ('Scalar shape',scalar.shape)
```

```python
#Vector
Vector = np.array([1,2,3])
print('Vector : ', Vector)
print ('Vector shape',Vector.shape)
```

```python
#Matrix
matrix=np.array([[1,2,3],[4,5,6]])
print('matrix : ', matrix)
print ('matrix shape',matrix.shape)
```

```python
#Tensor
tensor=np.array([[[1,2,3],[4,5,6]],[[78,9,4],[4,8,9]]])
print('tensor : ', tensor)
print ('tensor shape',tensor.shape)
```
