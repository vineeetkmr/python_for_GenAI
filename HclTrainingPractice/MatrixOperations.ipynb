```python
import numpy as np

a=np.array([[1,2,5,3],[3,5,6,7],[10,2,9,7],[7,5,0,1]])
b=np.array([[1,4,6,3],[8,7,9,7],[11,4,18,7],[8,3,9,4]])
print('a:',a)
print('Shape "a":',a.shape)
print('*'*10)

print('b:',b)
print('Shape "b"', b.shape)
```

```python
#addition
print('a+b :', a+b)
```

```python
# subtraction
print('a-b :', a-b)
```

```python
#multiplication - elementwise
print('a*b :', a*b)
```

```python
#dot product multiplication
print('a dot b', a@b)
```

```python
# dot product using numpy.dot
print('a dot b', np.dot(a,b))
```

```python
# matrix division
print('a/b :', a/b)
```

```python
#scalar multiplication
print(a*5)
```

```python
# matrix transpose
print('Transpose of a :', a.T)
```

```python
# matrix determinent
print('Determinant of a :', np.linalg.det(a))
```

```python
#matrix rank
print('Rank of a :', np.linalg.matrix_rank(a))
```

```python
#matrix inverse
print('Inverse of a :', np.linalg.inv(a))
```

```python
#Eigenvalues & EigenVectors
eigenvalues, eigenvectors = np.linalg.eig(a)
print('Eigenvalues :', eigenvalues)
print('Eigenvectors :', eigenvectors)
```

```python
#matrix concatenation (horizontal)
print('Horizontal Concatenation :', np.hstack((a,b)))
```

```python
#matrix concatenation (vertical)
print('Vertical Concatenation :', np.vstack((a,b)))
```

```python
#matrix reshape
print('Reshape of a :', a.reshape(4,4))

print('*'*5)

#or
print(np.linalg.norm(a))
```

```python
#Power (element wise power)
print("Power (element wise power)", a**2)
```
