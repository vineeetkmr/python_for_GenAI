```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#creating data frames
df=pd.DataFrame({
    "Month":['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'],
    "Temperature":[12,17,18,25,32,39,40,33,30,29,27,11]
})
```

```python
#data visualization in chart

plt.plot(df['Month'],df['Temperature'])

plt.title('Temperature in Different Months')
plt.xlabel('Month')
plt.ylabel('Temperature')

plt.show()
```
