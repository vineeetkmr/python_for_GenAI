```python
#importing libraries
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
# Task - 1. Load and display customer data
data=pd.read_csv("https://raw.githubusercontent.com/vineeetkmr/python_for_GenAI/refs/heads/master/myTestApp/HclTrainingPractice/Assignments/module2/customer_raw.csv")

#print('Customer data :\n',data)

# Display - head, info, describe
print('Head : \n',data.head())
print('*'*20)
print('Info:\n',data.info())
print('*'*20)
print('Describe:\n',data.describe(include="all"))
```

```python
# Task 2 Fix Missing Data

#A. Numerical Columns (Age, Income, SpendingScore)

# Convert Age to numeric (handle text later)
data['Age'] = data['Age'].replace({'twenty five':25}).astype(float)

# Clean Income
data['Income'] = data['Income'].astype(str)

data['Income'] = data['Income'].str.replace('₹','', regex=True)

data['Income'] = data['Income'].str.replace('k','000', regex=True)
data['Income'] = data['Income'].str.replace(',','', regex=True)

# Handle "nan" strings (from missing values)
data['Income'] = data['Income'].replace({'nan': np.nan})

#Convert to float
data['Income'] = data['Income'].astype(float)

```

```python
#Detect & Fix Outliers

# IQR method for SpendingScore
Q1 = data['SpendingScore'].quantile(0.25)
Q3 = data['SpendingScore'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

data['SpendingScore'] = np.where(data['SpendingScore']>upper, upper,
                        np.where(data['SpendingScore']<lower, lower, data['SpendingScore']))
```

```python
#Final Clean Dataset

data_cleaned = data.copy()
print(data_cleaned.info())
```
